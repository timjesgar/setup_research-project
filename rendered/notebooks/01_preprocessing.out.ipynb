# 1. Preprocessing

```

# load libs
library(tidyverse)
```

## Read Raw Data

```

df_raw <- read_csv(here('data/raw/raw_example.csv'))
```

## Preprocess Data

```

df_clean <-
  df_raw |>
  ## vs and and am are dichotomous -> change to logical values
  mutate(
    vs = as.logical(vs),
    am = as.logical(am)
  )

# examine cleaned data
summary(df_clean)
```

## Save Processed Data

```

write_csv(df_clean, here('data/processed/example.csv'))
```
