# 3. Plotting

```

# load libs
library(tidyverse)
```

## Read Data / Models

```

df <- read_csv(
  here('data/processed/example.csv'),
  show_col_types = FALSE
)

mod_lm <- readRDS(here('mods/example.rds'))
```

## Create Plots

```

# create plot
plot_mod <-
  ggplot(
    df,
    aes(
      x = hp,
      y = mpg,
      color = as.factor(gear)
    )
  ) +
  geom_point() +
  geom_abline(
    slope = mod_lm$coefficients['hp'],
    intercept = mod_lm$coefficients['(Intercept)']
  ) +
  labs(
    title = 'Predicted Relation of hp and mpg',
    color = 'Gear'
  )


# show plot
plot_mod
```

## Save Plots

```

ggsave(
  here('plots/example.png'),
  plot = plot_mod,
  width = custom_width,
  height = custom_height,
  units = custom_unit,
  dpi = custom_dpi
)
```
