# 2. Analysis

```

# load libs
library(tidyverse)
```

# Read Processed Data

```

df <-
  read_csv(
    here('data/processed/example.csv'),
    show_col_types = FALSE
  ) |>
  ## gear, is categorical -> define as factor
  mutate(
    gear = factor(gear)
  ) |>
  ## we only examine the effect of hp and gear on mpg
  select(mpg, hp, gear)
```

# Analysis

```

mod_lm <-
  lm(
    mpg ~ hp + gear,
    data = df
  )

# examine model
summary(mod_lm)
```

# Examine Model

```

mod_lm |>
  tidy() |>
  kable()
```

# Save Analysis/Model Objects

```

saveRDS(mod_lm, here('mods/example.rds'))
```
